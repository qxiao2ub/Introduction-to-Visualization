2. Basic Plots with Lines and/or Markers

```python
import matplotlib.pyplot as plt
```

```python
y = [31, 31, 41, 52, 63, 73, 79, 78, 70, 58, 48, 35]
x = list(range(1,13))

plt.plot(x,y)
plt.show()
```

2.1 Use makerks for the data points

Our first plot simply showed line segments. What if we want a scatter plot instead? Let's test out these commands individually:

```python
plt.plot(x, y, 'o')
plt.plot(x,y,'+')
plt.plot(x,y,'v')
```

```python
y1 = y
x1 = x
#plt.plot(x1, y1, 'rD') #red diamonds
#plt.plot(x1, y1, 'gH') #green hexagons
plt.plot(x1, y1, 'yo') #yellow dots
plt.show()
```

2.3 Change the size of the markers

```python
plt.plot(x1, y1, 'rD', markersize = 12)
```

```python
plt.plot(x1, y1, 'rD', ms = 14)
```

2.4 Draw the marker with a different outline color

Our markers can have outline and fill colors that are different.

Diamond markers filled with red and outlined in blue.

```python
plt.plot(x1, y1, 'rD', ms = 12, markeredgecolor= 'blue')
plt.plot(x1, y1, 'rD', ms = 14, mec = 'blue')
plt.plot(x1, y1, 'rD', ms = 10, markeredgecolor = 'blue', markeredgewidth = 4)
plt.plot(x1, y1, 'rD', ms = 12, mec='blue', mew = 5)
```

2.6 Color the line

We will now include lines connecting our markers. By default, the line and markers will be the same color. But, we can customize this.

Setting line and marker colors:

```python
# markers and line are the same color in red:
plt.plot(x1, y1, 'rD-', ms = 12)

# green hexagon markers and a red line:
plt.plot(x1, y1, 'H-', ms = 12, color = 'red', markerfacecolor = 'green')
plt.plot(x1, y1, 'H-', ms = 10, color = 'red', mfc = 'green')
```

2.7 Customizing the line style

Customized line styles, using dots as markers

```python
plt.plot(x1, y1, 'o--')
plt.plot(x1, y1, 'o', linestyle='dashed')
plt.plot(x1, y1, 'o', ls = 'dashed')
plt.plot(x1, y1, ls = 'dashed', marker = 'o')
plt.plot(x1, y1, 'o-')
```

```python
y2 = y
x2 = x
plt.plot(x2, y2, 'o-.')
plt.plot(x2, y2, 'o:')
```

2.8 Custom hex colors

A small collection of basic colors are provided. To really customize your plot you can use "hex" colors.

```python
plt.plot(x2, y2, 'o--', color = '#ff8080')
```

2.9 Add title to our plot

```python
plt.title('Plot 1')
```

2.10 Add axis labels

```python
plt.ylabel('Temperature')
plt.xlabel('Month')
```

2.11 Replace month numbers with names

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(x, months)
```

2.12 Add a second data series and a legend

Let's add the Miami high temperatures and a legend

```python
miami_high = [73, 74, 75, 78, 81, 84, 85, 86, 85, 82, 78, 76]

plt.plot(x, y , ls = 'dashed', marker = 'o', label = 'Buffalo')
plt.plot(x, miami_high, 'd', label = 'Miami')
plt.legend()
```
